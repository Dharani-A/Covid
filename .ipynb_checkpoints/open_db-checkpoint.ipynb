```python
#create a translator object
translator = Translator()

#Making copy of original dataset
df_chineseData = df_SimplifiedChinese.copy()
#print(df_chineseData['positive'])

#columns rename (not necassary)
df_chineseData.rename(columns=lambda x: translator.translate(x).text, inplace=True)

#translated column names
df_chineseData.columns

sample_column = ['gender','positive','fever', 'cough','smoke']

translations = {}
for column in sample_column:
    # unique elements of the column
    unique_elements = df_chineseData[column].unique()
    #print(unique_elements)
    for element in unique_elements:
        # add translation to the dictionary
        translations[element] = translator.translate(element).text
        
#print(translations)
#Replacing the values in the datframe with the translated data
df_chineseData.replace(translations, inplace = True)
```
